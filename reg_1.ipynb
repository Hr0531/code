```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
os.chdir("/Users/h-2/Documents")
df_hpa = pd.read_csv('hpa05.csv')
df_hpa.head()
```

```python
x = df_hpa[['Fukuoka']]
y = df_hpa[['Tokyo']]

plt.scatter(x, y)
```

```python
model = LinearRegression()
model.fit(x, y)
print('coefficient = ', model.coef_[0])
print('intercept = ', model.intercept_)
```

```python
plt.scatter(x, y)
plt.plot(x, model.predict(x))
```
