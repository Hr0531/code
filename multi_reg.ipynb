```python
import os
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
os.chdir("/Users/h-2/Documents")
df = pd.read_csv('hpa05_2.csv')
df.head()
```

```python
df.corr()
```

```python
sns.pairplot(df, size = 1.0)
```

```python
x = df[['Fukuoka', 'Fukuoka_Temp']].values
y = df[['Tokyo']].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 0)

model_multi = LinearRegression()

model_multi.fit(x_train, y_train)

y_train_predict = model_multi.predict(x_train)

y_test_predict = model_multi.predict(x_test)

df_coefficient = pd.DataFrame(model_multi.coef_.reshape(1, 2), columns = ['Fukuoka', 'Fukuoka_Temp'])
print(df_coefficient)
```

```python
print('intercept: %.3f' % model_multi.intercept_)
```
